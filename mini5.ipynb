```python
# To-Do List Application

# Dictionary to store tasks
tasks = {
    "Task1":"wake up",
    "Task2":"reading",
    "Task3":"watching mobile",
    "Task4":"finish washings",
    "Task5":"finishing breakfast",
    "Task6":"completing the work"
}

def create_task(task_id, description):
    if task_id in tasks:
        return f"Task ID {task_id} already exists."
    tasks[task_id] = description
    return f"Task '{description}' created with ID {task_id}."

def update_task(task_id, new_description):
    if task_id not in tasks:
        return f"Task ID {task_id} does not exist."
    tasks[task_id] = new_description
    return f"Task ID {task_id} updated to '{new_description}'."

def delete_task(task_id):
    if task_id not in tasks:
        return f"Task ID {task_id} does not exist."
    deleted_task = tasks.pop(task_id)
    return f"Task ID {task_id} ('{deleted_task}') deleted."

def view_tasks():
    if not tasks:
        return "No tasks available."
    return "\n".join([f"ID: {id}, Task: {desc}" for id, desc in tasks.items()])

# Main function to handle user commands
def main():
    print("Welcome to the To-Do List Application!")
    print("Commands: create, update, delete, view, exit")
    
    # Get user command
    command = input("Enter command: ").strip().lower()

    # Use match-case to handle commands
    match command:
        case "create":
            task_id = input("Enter task ID: ")
            description = input("Enter task description: ")
            print(create_task(task_id, description))
            
        case "update":
            task_id = input("Enter task ID: ")
            new_description = input("Enter new task description: ")
            print(update_task(task_id, new_description))
            
        case "delete":
            task_id = input("Enter task ID: ")
            print(delete_task(task_id))
            
        case "view":
            print(view_tasks())
            
        case "exit":
            print("Exiting the application.")
            
        case _:
            print("Invalid command. Please try again.")
            
            if __name__=="__main__":
                main()

    
```
